# **Questão 1 - C**

```python
def fatComRepeticao(N):
  output = 1

  for i in range(2, N + 1):
    output*=i

  return output


def fatComRecurcao(N):
  if N == 1:
    return 1

  return N*fatComRecurcao(N-1)


print('Fatorial utilizando repetição: ',fatComRepeticao(4))
print('Fatorial utilizando recursão: ',fatComRecurcao(4))
```

# **Questão 1 - G**

```python
def quantizacaoRGB(image, bits):
    N = image.shape[0]
    M = image.shape[1]
    output = np.array([[[0]*3]*M]*N)

    for i in range(N):
      for j in range(M):
        for k in range(3):
          # Processo de quantização:
          output[i][j][k] = round((image[i][j][k]/(2**8 - 1))*((2**bits)-1))
          # Processo de normalização
          output[i][j][k] = output[i][j][k]/((2**bits)-1)*(2**8 - 1)


    return output

im_cor = mpimage.imread("lena_cor.bmp")

print("QUANTIZAÇÃO DA IMAGEM RGB")
print("8 Bits:")
output = quantizacaoRGB(im_cor, 8)
imshow(output), show()
print('\n')

print("7 Bits:")
output = quantizacaoRGB(im_cor, 7)
imshow(output), show()
print('\n')

print("6 Bits:")
output = quantizacaoRGB(im_cor, 6)
imshow(output), show()
print('\n')

print("5 Bits:")
output = quantizacaoRGB(im_cor, 5)
imshow(output), show()
print('\n')

print("4 Bits:")
output = quantizacaoRGB(im_cor, 4)
imshow(output), show()
print('\n')

print("3 Bits:")
output = quantizacaoRGB(im_cor, 3)
imshow(output), show()
print('\n')

print("2 Bits:")
output = quantizacaoRGB(im_cor, 2)
imshow(output), show()
print('\n')

print("1 Bits:")
output = quantizacaoRGB(im_cor, 1)
imshow(output), show()
print('\n')
```

```python
def quantizacaoGray(image, bits):
    N = image.shape[0]
    M = image.shape[1]
    output = np.array([[0]*M]*N)

    for i in range(N):
      for j in range(M):
          # Processo de quantização
          output[i][j] = round((image[i][j]/(2**8 - 1))*((2**bits)-1))
          # Processo de normalização
          output[i][j] = (output[i][j]/((2**bits)-1))*(2**8 - 1)

    return output

im_gray = mpimage.imread("lena_gray.bmp")

print("QUANTIZAÇÃO DA IMAGEM RGB")
print("8 Bits:")
output = quantizacaoGray(im_gray, 8)
imshow(output, cmap='gray'), show()
print('\n')

print("7 Bits:")
output = quantizacaoGray(im_gray, 7)
imshow(output, cmap='gray'), show()
print('\n')

print("6 Bits:")
output = quantizacaoGray(im_gray, 6)
imshow(output, cmap='gray'), show()
print('\n')

print("5 Bits:")
output = quantizacaoGray(im_gray, 5)
imshow(output, cmap='gray'), show()
print('\n')

print("4 Bits:")
output = quantizacaoGray(im_gray, 4)
imshow(output, cmap='gray'), show()
print('\n')

print("3 Bits:")
output = quantizacaoGray(im_gray, 3)
imshow(output, cmap='gray'), show()
print('\n')

print("2 Bits:")
output = quantizacaoGray(im_gray, 2)
imshow(output, cmap='gray'), show()
print('\n')

print("1 Bits:")
output = quantizacaoGray(im_gray, 1)
imshow(output, cmap='gray'), show()
print('\n')
```
